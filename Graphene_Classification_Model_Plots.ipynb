```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_confusion_matrix(tp, fp, fn, tn, title, color):
    # Create the confusion matrix array
    matrix = np.array([[tp, fp], [fn, tn]])

    # Calculate total predictions
    total = tp + fp + fn + tn

    # Calculate the percentages
    percentages = matrix / total * 100

    # Labels for each box
    labels = np.array([[f'TP\n{tp}\n{percentages[0, 0]:.1f}%',
                        f'FP\n{fp}\n{percentages[0, 1]:.1f}%'],
                       [f'FN\n{fn}\n{percentages[1, 0]:.1f}%',
                        f'TN\n{tn}\n{percentages[1, 1]:.1f}%']])

    # Plot the confusion matrix with background shading
    plt.figure(figsize=(6, 5))
    sns.heatmap(percentages, annot=labels, fmt='', cmap=color, cbar=False, square=True,
                annot_kws={'size': 14, 'weight': 'bold'}, linewidths=0.5, linecolor='black')

    # Set axis labels and title
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.xticks([0.5, 1.5], ["Positive", "Negative"])
    plt.yticks([0.5, 1.5], ["Positive", "Negative"], rotation=0)
    plt.title(title)
    plt.show()

tp, fp, fn, tn = 40, 10, 1, 49
plot_confusion_matrix(tp, fp, fn, tn, "ResNet50 Confusion Matrix", 'Greens')
```
